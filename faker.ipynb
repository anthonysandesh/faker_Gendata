```python
from faker import Faker
fake = Faker()
```

```python
ssn = []
credit_card = []
```

```python
#SSN
Faker.seed(0)
for _ in range(100000):
    ssn.append(fake.ssn())
```

```python
# bank account
Faker.seed(0)
for _ in range(5):
    print(fake.iban())
```

```python
#credit card
Faker.seed(0)
for _ in range(100000):
    credit_card.append(fake.credit_card_number(card_type='visa'))
    
```

```python
# passport number
Faker.seed(0)
for _ in range(5):
    print(fake.passport_number())
```
